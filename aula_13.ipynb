```python
# Passo 1 - Importar pandas
import pandas as pd

# Passo 2 - Ler arquivo

dados = pd.read_csv('Soccer_Football Clubs Ranking.csv')
dados
```

```python
# Passo 4 - Criar Filtro
filtro_por_country = data_frame['country'] == ''
dados_filtrados = data_frame[filtro_por_country].copy()
```
